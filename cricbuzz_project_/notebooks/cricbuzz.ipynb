```python
import requests
from datetime import datetime

# Step 1: API Setup
BASE_URL = "https://cricbuzz-cricket.p.rapidapi.com"
HEADERS = {
    "x-rapidapi-key": "f9d01a5f6fmsh3162b874a7b79ffp1f155bjsna4b17da5fe78",
    "x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

# Step 2: Get Live Matches
url = f"{BASE_URL}/matches/v1/live"
response = requests.get(url, headers=HEADERS)

# Check if response is OK
if response.status_code != 200:
    print(f"‚ö†Ô∏è API Request Failed. Status Code: {response.status_code}")
    print(response.text)
else:
    data_live = response.json()

    matches_live = []
    for type_match in data_live.get('typeMatches', []):
        for series in type_match.get('seriesMatches', []):
            series_data = series.get('seriesAdWrapper', {})
            series_id = series_data.get('seriesId')
            series_name = series_data.get('seriesName')

            for match in series_data.get('matches', []):
                match_info = match.get('matchInfo', {})
                start_date_raw = match_info.get('startDate')
                end_date_raw = match_info.get('endDate')

                start_date = (
                    datetime.fromtimestamp(int(start_date_raw) / 1000).strftime("%Y-%m-%d %H:%M:%S")
                    if start_date_raw else None
                )
                end_date = (
                    datetime.fromtimestamp(int(end_date_raw) / 1000).strftime("%Y-%m-%d %H:%M:%S")
                    if end_date_raw else None
                )

                matches_live.append({
                    'series_id': series_id,
                    'series_name': series_name,
                    'match_id': match_info.get('matchId'),
                    'match_desc': match_info.get('matchDesc'),
                    'match_format': match_info.get('matchFormat'),
                    'start_date': start_date,
                    'end_date': end_date,
                    'team1': match_info.get('team1', {}).get('teamName'),
                    'team2': match_info.get('team2', {}).get('teamName'),
                    'venue': match_info.get('venueInfo', {}).get('ground')
                })

# ‚úÖ Display all live matches
if matches_live:
    for match in matches_live:
        print(f"""
Match ID: {match.get('match_id', 'N/A')}
Series: {match.get('series_name', 'N/A')}
Match: {match.get('match_desc', 'N/A')} ({match.get('match_format', 'N/A')})
Teams: {match.get('team1', 'N/A')} vs {match.get('team2', 'N/A')}
Venue: {match.get('venue', 'N/A')}
Start: {match.get('start_date', 'N/A')}
End:   {match.get('end_date', 'N/A')}
--------------------------
""")
else:
    print("‚ö†Ô∏è No live matches found.")

```

==== Extract All live matches and their score from the API by using nested for loops ====

```python
# ==== Extract All Live Matches and Their Score ====
import requests
import pandas as pd
from datetime import datetime

# Step 1: API Setup
BASE_URL = "https://cricbuzz-cricket.p.rapidapi.com"
HEADERS = {
    "x-rapidapi-key": "f9d01a5f6fmsh3162b874a7b79ffp1f155bjsna4b17da5fe78",  # Replace with your key
    "x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

# Step 2: Get Live Matches
url = f"{BASE_URL}/matches/v1/live"
response = requests.get(url, headers=HEADERS)

if response.status_code != 200:
    print(f"‚ö†Ô∏è Failed to fetch matches. Status Code: {response.status_code}")
    exit()

data_live = response.json()
matches_live = []

# Step 3: Extract Live Match Info
for type_match in data_live.get('typeMatches', []):
    for series in type_match.get('seriesMatches', []):
        series_data = series.get('seriesAdWrapper', {})
        series_name = series_data.get('seriesName')

        for match in series_data.get('matches', []):
            match_info = match.get('matchInfo', {})
            match_id = match_info.get('matchId')

            start_date = datetime.fromtimestamp(int(match_info['startDate']) / 1000).strftime("%Y-%m-%d %H:%M:%S") if match_info.get('startDate') else None
            end_date = datetime.fromtimestamp(int(match_info['endDate']) / 1000).strftime("%Y-%m-%d %H:%M:%S") if match_info.get('endDate') else None

            matches_live.append({
                'match_id': match_id,
                'series_name': series_name,
                'match_desc': match_info.get('matchDesc'),
                'match_format': match_info.get('matchFormat'),
                'start_date': start_date,
                'end_date': end_date,
                'team1': match_info.get('team1', {}).get('teamName'),
                'team2': match_info.get('team2', {}).get('teamName'),
                'venue': match_info.get('venueInfo', {}).get('ground')
            })

# Step 4: Display Match Info & Fetch Scorecard
for match in matches_live:
    print(f"""
üì¢ Series: {match['series_name']}
üìÖ Match: {match['match_desc']} ({match['match_format']})
üèüÔ∏è Venue: {match['venue']}
üèè Teams: {match['team1']} vs {match['team2']}
‚è±Ô∏è Start: {match['start_date']}
‚è±Ô∏è End:   {match['end_date']}
--------------------------
    """)

    #Scorecard Endpoint
    score_url = f"{BASE_URL}/mcenter/v1/{match['match_id']}/scard"
    score_response = requests.get(score_url, headers=HEADERS)

    if score_response.status_code == 200:
        score_data = score_response.json()

        if "scorecard" in score_data:
            for i, innings in enumerate(score_data["scorecard"], start=1):
                bat_team = innings.get("batteamname", "Unknown Team")
                print(f"üìä Inning {i}: {bat_team}")

                # üèè Batting Table
                batting_df = pd.DataFrame(innings.get("batsman", []))
                if not batting_df.empty:
                    batting_df = batting_df[['name', 'runs', 'balls', 'fours', 'sixes', 'strkrate', 'outdec']]
                    batting_df.rename(columns={
                        'name': 'Batsman',
                        'runs': 'Runs',
                        'balls': 'Balls',
                        'fours': '4s',
                        'sixes': '6s',
                        'strkrate': 'SR',
                        'outdec': 'How Out'
                    }, inplace=True)
                    print("\nüèè Batting Scorecard:")
                    print(batting_df.to_string(index=False))
                else:
                    print("No batting data available.")

                # üéØ Bowling Table
                bowling_df = pd.DataFrame(innings.get("bowler", []))
                if not bowling_df.empty:
                    bowling_df = bowling_df[['name', 'overs', 'maidens', 'runs', 'wickets', 'economy']]
                    bowling_df.rename(columns={
                        'name': 'Bowler',
                        'overs': 'Overs',
                        'maidens': 'Maidens',
                        'runs': 'Runs',
                        'wickets': 'Wkts',
                        'economy': 'Econ'
                    }, inplace=True)
                    print("\nüéØ Bowling Scorecard:")
                    print(bowling_df.to_string(index=False))
                else:
                    print("No bowling data available.")

                print("\n---------------------------------\n")
        else:
            print("‚ö† No scorecard available for this match.\n")
    else:
        print(f"‚ö† Could not fetch scorecard (Error {score_response.status_code})\n")
```

```python
# ======================================================
# üèè Cricbuzz Player Stats + ICC Rankings (Step by Step)
# ======================================================

import requests
import pandas as pd
from urllib.parse import quote

# ------------------ STEP 1: Setup API ------------------
API_KEY = "f9d01a5f6fmsh3162b874a7b79ffp1f155bjsna4b17da5fe78"  # Replace with your key if needed
HEADERS = {
    "x-rapidapi-key": API_KEY,
    "x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}
BASE_URL = "https://cricbuzz-cricket.p.rapidapi.com"
```

```python
# ------------------ STEP 2: Search Player ------------------
def search_player(query):
    """
    Search for a player by name and return the first match.
    """
    url = f"{BASE_URL}/stats/v1/player/search?plrN={quote(query)}"
    r = requests.get(url, headers=HEADERS, timeout=10)
    r.raise_for_status()
    players = r.json().get("player", [])
    return players[0] if players else None

result = search_player("Virat Kohli")
print(result)
```

```python
# ------------------ STEP 3: Fetch Batting / Bowling Stats ------------------
def fetch_stats(player_id, stat_type):
    url = f"{BASE_URL}/stats/v1/player/{player_id}/{stat_type}"
    r = requests.get(url, headers=HEADERS, timeout=10)
    if r.status_code == 404:
        return pd.DataFrame()  # No data available
    r.raise_for_status()
    data = r.json()
    if "headers" in data and "values" in data:
        headers = data["headers"]
        rows = [row["values"] for row in data["values"]]
        df = pd.DataFrame(rows, columns=headers)
        return df.dropna(axis=1, how="all")  # Clean empty columns
    return pd.DataFrame()
# Fetch batting stats
bat_df = fetch_stats(1413, "batting")
print("üìä Batting Stats:")
if not bat_df.empty:
    print(bat_df.to_string(index=False))
else:
    print("No batting data available.\n")

# Fetch bowling stats
bowl_df = fetch_stats(1413, "bowling")
print("üéØ Bowling Stats:")
if not bowl_df.empty:
    print(bowl_df.to_string(index=False))
else:
    print("No bowling data available.\n")

```

```python
#==================================================
# üèè Cricbuzz Player Stats + ICC Rankings 
#==================================================



import requests
import pandas as pd
from urllib.parse import quote

API_KEY = "fb04535bebmsh9d58b5cd1fe8dd6p14e20djsn10e48abe5bad"
HEADERS = {"x-rapidapi-key": API_KEY, "x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"}
BASE_URL = "https://cricbuzz-cricket.p.rapidapi.com"

def search_player(query):
    url = f"{BASE_URL}/stats/v1/player/search?plrN={quote(query)}"
    r = requests.get(url, headers=HEADERS, timeout=10)
    r.raise_for_status()
    players = r.json().get("player", [])
    return players[0] if players else None

def fetch_stats(player_id, stat_type):
    url = f"{BASE_URL}/stats/v1/player/{player_id}/{stat_type}"
    r = requests.get(url, headers=HEADERS, timeout=10)
    if r.status_code == 404:
        return pd.DataFrame()
    r.raise_for_status()
    data = r.json()
    if "headers" in data and "values" in data:
        headers = data["headers"]
        rows = [row["values"] for row in data["values"]]
        df = pd.DataFrame(rows, columns=headers)
        return df.dropna(axis=1, how="all")  # clean empty columns
    return pd.DataFrame()

def fetch_player_rankings(player_id, player_name):
    # ---- Try Player-specific rankings API ----
    url = f"{BASE_URL}/stats/v1/player/{player_id}/iccRankings"
    r = requests.get(url, headers=HEADERS, timeout=10)
    rows = []

    if r.status_code == 200:
        data = r.json()
        if "rankings" in data and data["rankings"]:
            df = pd.DataFrame(data["rankings"])
            return sort_rankings(df)

    # ---- Fallback: Search in global ICC ranking tables ----
    formats = ["test", "odi", "t20i"]
    categories = ["batting", "bowling", "allrounder"]
    norm_name = player_name.lower().replace(" ", "")

    for fmt in formats:
        for cat in categories:
            url = f"{BASE_URL}/stats/v1/icc-rankings/{fmt}/{cat}"
            r = requests.get(url, headers=HEADERS, timeout=10)
            if r.status_code != 200:
                continue
            for p in r.json().get("rank", []):
                pname = p.get("name", "")
                if pname and pname.lower().replace(" ", "") == norm_name:
                    rows.append({
                        "format": fmt.upper(),
                        "category": cat.capitalize(),
                        "rank": p.get("rank"),
                        "rating": p.get("points")
                    })

    if not rows:
        return pd.DataFrame()
    return sort_rankings(pd.DataFrame(rows))

def sort_rankings(df):
    format_order = {"TEST": 1, "ODI": 2, "T20I": 3}
    category_order = {"Batting": 1, "Bowling": 2, "Allrounder": 3}
    df["fmt_sort"] = df["format"].map(format_order)
    df["cat_sort"] = df["category"].map(category_order)
    df = df.sort_values(["fmt_sort", "cat_sort"]).drop(columns=["fmt_sort", "cat_sort"])
    return df[["format", "category", "rank", "rating"]]

# ---------- MAIN ----------
if __name__ == "__main__":
    name = input("Enter player name: ").strip()
    if not name:
        print("‚ö† Please enter a valid name.")
    else:
        player = search_player(name)
        if not player:
            print("‚ùå No player found.")
        else:
            pid = player["id"]
            pname = player["name"]
            print(f"\n‚úÖ Player Found: {pname} ({player.get('teamName','N/A')})")
            print(f"üÜî Player ID: {pid}")

            print("\nüìä Batting Summary:")
            bat_df = fetch_stats(pid, "batting")
            print(bat_df.to_string(index=False) if not bat_df.empty else "No batting data available.")

            print("\nüéØ Bowling Summary:")
            bowl_df = fetch_stats(pid, "bowling")
            print(bowl_df.to_string(index=False) if not bowl_df.empty else "No bowling data available (Player rarely bowls).")

            print("\nüèÜ ICC Rankings:")
            icc_df = fetch_player_rankings(pid, pname)
            print(icc_df.to_string(index=False) if not icc_df.empty else "No ICC ranking data available.")
```
